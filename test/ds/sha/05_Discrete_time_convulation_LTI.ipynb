```python
import numpy as np
import matplotlib.pyplot as plt

# Length for visualization
N = 20

# Define x[n] = u[n]
x = np.ones(N)

# Define h[n] = u[n] - u[n - 5] â†’ a pulse of 5 ones
h = np.concatenate((np.ones(5), np.zeros(N - 5)))

# Compute convolution y[n] = x[n] * h[n]
y = np.convolve(x, h)

# Time indices
n_x = np.arange(N)
n_h = np.arange(N)
n_y = np.arange(len(y))

# Plotting
plt.stem(n_y, y)
plt.title('Output y[n] = x[n] * h[n]')
plt.xlabel('n')
plt.ylabel('y[n]')
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

def unit_step(n):
    return np.where(n>=0,1,0)

def manual_convo(x,h):
    N = len(x)
    M = len(h)
    y = np.zeros(N+M-1)# the function np.zeros(shape) creates a new array filled entirely with zeros.

    for n in range(len(y)):
        for k in range(N):
            if 0<= n-k<M:
                y[n]+=x[k]*h[n-k]
    return y

n = np.arange(0,21)
x = unit_step(n)

h = unit_step(n) - unit_step(n-5)

y = manual_convo(x,h)

n_y = np.arange(0,21)
y = y[:21]

plt.figure(figsize=(14,6))

plt.subplot(3,1,1)
plt.stem(n,h,label = 'unit step')
plt.title("input x[n] = u(n)")
plt.xlabel("n")
plt.ylabel('x[n]')
plt.legend()
plt.grid()

plt.subplot(3,1,2)
plt.stem(n,x,label = 'impulse response')
plt.title("impulse h[n] = u(n) - u(n-5)")
plt.xlabel("n")
plt.ylabel('x[n]')
plt.legend()
plt.grid()

plt.subplot(3,1,3)
plt.stem(n_y,y,label = 'convolution')
plt.title("output y[n] = x[n]*h[n]")
plt.xlabel("n")
plt.ylabel('y[n]')
plt.legend()
plt.grid()

plt.tight_layout()
plt.show()
```
