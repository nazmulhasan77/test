```python
import numpy as np
import matplotlib.pyplot as plt

n = np.arange(0,11,1)

omega1 = 0.25*np.pi
omega2 = 0.5*np.pi
omega3 = np.pi

x1 = np.cos(omega1*n)
x2 = np.cos(omega2*n)
x3 = np.cos(omega3*n)

plt.figure(figsize=(25,15))

plt.subplot(3,1,1)
plt.stem(n,x1,label='w=pi/4')
plt.title("Low oscillation at w = pi/4")
plt.xlabel('n')
plt.ylabel("x[n]")
plt.legend()

plt.subplot(3,1,2)
plt.stem(n,x2,label="w=pi/2")
plt.title("Low oscillation at w = PI/2")
plt.xlabel("n")
plt.ylabel("x[n]")
plt.legend()

plt.subplot(3,1,3)
plt.stem(n,x3,label = 'w=PI')
plt.title("Maximum oscillation at w = PI")
plt.xlabel("n")
plt.ylabel("x[n]")
plt.legend()

plt.show()
```
