```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
n = np.arange(-10,10,1)
u_n = np.where(n>=0,1,0)

plt.stem(n,u_n,label="Unit Step Sequence")
plt.xlabel("n",color = "blue")
plt.ylabel("U_n",color = "blue")
plt.legend()
plt.grid()

plt.show()
```

```python
n = np.arange(-10,10,1)
r_n = np.where(n>=0,n,0)

plt.stem(n,r_n,label="Ramp Sequence")
plt.xlabel("n",color = "red")
plt.ylabel("r_n",color = "red")
plt.legend()
plt.grid()

plt.show()
```

```python

A = 1
b = 0.5
t = np.linspace(0,10,50)
y = A*np.exp(b*t)


plt.stem(t,y,label="exponential growth")
plt.title("Exponential Function")
plt.xlabel("Time(t)",color = "red")
plt.ylabel("y(n)",color = "red")
plt.legend()
plt.grid()

plt.show()
```

```python

A = 1
f = 1
t = np.linspace(0,2,500)
y = A*np.sin(2*np.pi*f*t)


plt.plot(t,y,label="Sine Wave")
plt.title("Sine Wave")
plt.xlabel("Time(s)",color = "red")
plt.ylabel("Amplitude",color = "red")
plt.axvline(x=0,color = 'blue')
plt.axhline(y=0,color = "Blue")
plt.legend()
plt.grid()

plt.show()
```

```python
A = 1
f = 1
t = np.linspace(0,2,500)
phi = np.pi/2
y = A*np.cos(2*np.pi*f*t+phi)

plt.plot(t,y,label = "Cosine")
plt.title("Cosine Wave")
plt.xlabel("time(s)",color = 'orange')
plt.ylabel("Amplitude",color = 'orange')
plt.grid()
plt.legend()
plt.axvline(x=0,color='black')
plt.axhline(y = 0,color = 'blue')

plt.show
```
