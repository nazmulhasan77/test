```python
import numpy as np
import matplotlib.pyplot as plt

x =np.array([1,3,-2,4])
y = np.array([2,3,-1,3])
z = np.array([2,-1,4,-2])

def cross_correlation(x,y):
    N = len(x)
    M = len(y)
    corr_len = N+M-1
    result = np.zeros(corr_len)
    x_padded = np.pad(x,(M-1,M-1))

    for i in range(corr_len):
        for j in range(M):
            result[i]+=y[j] *x_padded[i+j]
    return result

#r_xy = np.correlate(x,y,mode='full')
#r_yz = np.correlate(y,z,mode='full')

r_xy = cross_correlation(x,y)
r_yz = cross_correlation(y,z)
lags = np.arange(-len(x)+1,len(x))

plt.figure(figsize=(12, 5))

# Plot r_xy
plt.subplot(1, 2, 1)
plt.stem(lags, r_xy, basefmt=" ")
plt.title("Cross-Correlation: x(n) vs y(n)")
plt.xlabel("Lag")
plt.axvline(color='red')
plt.axhline(color='red')
plt.ylabel("Correlation")
plt.grid(True)

# Plot r_yz
plt.subplot(1, 2, 2)
plt.stem(lags, r_yz, basefmt=" ")
plt.title("Cross-Correlation: y(n) vs z(n)")
plt.xlabel("Lag")
plt.ylabel("Correlation")
plt.axvline(color='red')
plt.axhline(color='red')
plt.grid(True)

plt.tight_layout()
plt.show()
```

```python
x = np.array([1,3,-2,4])
y = np.array([2,3,-1,3])
z = np.array([2,-1,4,-2])

#r_xy = np.correlate(x,y,mode = 'full')
#r_yz = np.correlate(y,z,mode = 'full')
def manu_correlation(x,y):
    N = len(x)
    M = len(y)
    result = np.zeros(N+M-1)
    x_padded = np.pad(x,(M-1,M-1))

    for i in range(M+N-1):
        for j in range(M):
            result[i]+=y[j]*x_padded[i+j]
    return result


r_xy = manu_correlation(x,y)
r_yz = manu_correlation(y,z)

lags = np.arange(-len(x)+1,len(x))

plt.figure(figsize=(20,8))

plt.subplot(2,1,1)
plt.stem(lags,r_xy,label = "correlation of x&y")
plt.title("Correlation between x and y")
plt.xlabel("Lag")
plt.axvline(color ='red')
plt.axhline(color = 'red')
plt.ylabel("Correaltion")
plt.legend()
plt.grid()


plt.subplot(2,1,2)
plt.stem(lags,r_yz,label = "correlation of y&z")
plt.title("Correlation between y and z")
plt.xlabel("Lag")
plt.ylabel("Correaltion")
plt.axvline(color ='red')
plt.axhline(color = 'red')
plt.legend()
plt.grid()

plt.tight_layout()
plt.show()
```
