```python
import numpy as np
import matplotlib.pyplot as plt

A = 3
f = 50
t_cont = np.linspace(0,0.08,1000)
x_cont = A*np.cos(2*np.pi*f*t_cont)

fs1 = 200
Ts1 = 1/fs1
n1 = np.arange(0,0.08,Ts1)
X_sampled_200Hz = A * np.cos(2*np.pi*f*n1)

fs2 = 75
ts2 = 1/fs2
n2 = np.arange(0,0.08,ts2)
X_sampled_75hz = A*np.cos(2*np.pi*f*n2)

plt.figure(figsize=(20,8))

plt.subplot(3,1,1)
plt.plot(t_cont,x_cont,label = 'Original Signal')
plt.title("Original signal: x(t)=3cos(100*PI*t)")
plt.xlabel("time(s)")
plt.ylabel('Amplitude')
plt.grid
plt.legend()

plt.subplot(3,1,2)
plt.plot(t_cont,x_cont,color = 'lightgray',label = 'Original Signal')
plt.stem(n1,X_sampled_200Hz,label = 'sampled signal')
plt.plot(n1,X_sampled_200Hz,label = 'sampled signal')
plt.title("Sampling at 200 Hz (No Aliasing)")
plt.xlabel('time(s)')
plt.ylabel('Amplitude')
plt.grid()
plt.legend()

plt.subplot(3,1,3)
plt.plot(t_cont,x_cont,color = 'lightgray',label = 'Original Signal')
plt.stem(n2,X_sampled_75hz,label = 'aliasing signal')
plt.plot(n2,X_sampled_75hz,label = 'aliasing signal')
plt.title("Sampled at 75hz (Aliasing)")
plt.xlabel("time(s)")
plt.ylabel("Amplitude")
plt.legend()
plt.grid()

plt.tight_layout()
plt.show()

```

```python
import numpy as np
import matplotlib.pyplot as plt

# Define the analog signal
def xa(t):
    return 3 * np.cos(2000 * np.pi * t) + 5 * np.sin(6000 * np.pi * t) + 10 * np.cos(12000 * np.pi * t)

# Continuous time for reference
t_cont = np.linspace(0, 0.002, 1000)
x_cont = xa(t_cont)

# Sampling at 16000 Hz
fs1 = 16000
t1 = np.arange(0, 0.002, 1/fs1)
x1 = xa(t1)

# Sampling at 12000 Hz
fs2 = 12000
t2 = np.arange(0, 0.002, 1/fs2)
x2 = xa(t2)

# Sampling at 8000 Hz
fs3 = 8000
t3 = np.arange(0, 0.002, 1/fs3)
x3 = xa(t3)

# Plotting
plt.figure(figsize=(12, 9))

plt.subplot(3, 1, 1)
plt.plot(t_cont , x_cont, color = 'lightgray', label='Original Signal')
plt.stem(t1 , x1, label='Sampled @ 16kHz')
plt.title('Sampling at 16000 Hz')
plt.xlabel('Time (ms)')
plt.ylabel('Amplitude')
plt.legend()
plt.grid(True)

plt.subplot(3, 1, 2)
plt.plot(t_cont , x_cont, color = 'lightgray', label='Original Signal')
plt.stem(t2 , x2, label='Sampled @ 12kHz')
plt.title('Sampling at 12000 Hz')
plt.xlabel('Time (ms)')
plt.ylabel('Amplitude')
plt.legend()
plt.grid(True)

plt.subplot(3, 1, 3)
plt.plot(t_cont , x_cont, color = 'lightgray', label='Original Signal')
plt.stem(t3 , x3, label='Sampled @ 8kHz')
plt.title('Sampling at 8000 Hz')
plt.xlabel('Time (ms)')
plt.ylabel('Amplitude')
plt.legend()
plt.grid(True)

plt.tight_layout()
plt.show()
```
