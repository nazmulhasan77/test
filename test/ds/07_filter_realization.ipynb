## ‚úÖ 1. **6-Point Averaging Filter**

### üìå Purpose:

Averaging filters are **low-pass filters** ‚Äî they smooth the input by averaging neighboring samples.

### üìò Difference Equation:

For a 6-point moving average:

$$
y[n] = \frac{1}{6} \left( x[n] + x[n-1] + x[n-2] + x[n-3] + x[n-4] + x[n-5] \right)
$$

This is a **FIR (Finite Impulse Response)** filter with coefficients:

$$
h[n] = \begin{cases}
\frac{1}{6} & \text{for } n = 0 \text{ to } 5 \\
0 & \text{otherwise}
\end{cases}
$$

### üîß Realization Structure:

You implement this using:

* 5 delay elements (`z‚Åª¬π`)
* 6 multipliers (each with gain $\frac{1}{6}$)
* 5 adders

It looks like a **tap-delay line** (direct form FIR):

## ‚úÖ 2. **6-Point Differencing Filter**

### üìå Purpose:

Differencing filters are **high-pass filters** ‚Äî they highlight **changes** in signal, such as edges.

### üìò Difference Equation:

A simple 6-point differencing filter can be:

$$
y[n] = x[n] - x[n-6]
$$

This captures the **difference between current and 6-step past input**.

Or, a generalized differencing version (finite difference with equal weights):

$$
y[n] = x[n] - x[n-1] + x[n-2] - x[n-3] + x[n-4] - x[n-5]
$$

### üîß Realization Structure:

Use:

* 5 delay elements
* 6 multipliers (+1 or -1)
* 5 adders

For example, for:

$$
y[n] = x[n] - x[n-6]
$$

Structure:

```
x[n] ‚îÄ‚î¨‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚ñ∂(+)‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚ñ∂ y[n]
      ‚îÇ
     z‚Åª‚Å∂
      ‚îÇ
      ‚îî‚îÄ‚îÄ‚îÄ‚îÄ‚ñ∂(-1)
```

For:

$$
y[n] = x[n] - x[n-1] + x[n-2] - x[n-3] + x[n-4] - x[n-5]
$$

You alternate signs and add up 6 delayed values.

---

## üß† Summary Table

| Filter Type       | Equation                                    | Behavior                   |
| ----------------- | ------------------------------------------- | -------------------------- |
| 6-pt Averaging    | $y[n] = \frac{1}{6} \sum_{k=0}^{5} x[n-k]$  | Low-pass (smooth)          |
| 6-pt Differencing | $y[n] = x[n] - x[n-6]$ or alternating signs | High-pass (detect changes) |

---

````python
import numpy as np
import matplotlib.pyplot as plt

# Create a simple signal with a step and a spike
x = [1,10,3,4,5,6,12,8,9,15]
n = np.arange(len(x))

# --- Low-pass filter: 6-point average ---
avg_filter = np.ones(6) / 6
low_pass = np.convolve(x, avg_filter, mode='same')  # smooth signal

# --- High-pass filter: difference with 6 steps before ---
x_padded = np.concatenate((np.zeros(6), x))         # pad beginning with 6 zeros
high_pass = x_padded[6:] - x_padded[:-6]            # x[n] - x[n-6]
high_pass = high_pass[:len(x)]                      # trim to original length

# --- Plot the results ---
plt.figure(figsize=(8, 6))

plt.subplot(3, 1, 1)
plt.stem(n, x)
plt.title("Original Signal")

plt.subplot(3, 1, 2)
plt.stem(n, low_pass)
plt.title("Low-Pass Filter (6-point Average)")

plt.subplot(3, 1, 3)
plt.stem(n, high_pass)
plt.title("High-Pass Filter (6-step Difference)")
print("averaging",[round(val,2) for val in low_pass])
plt.tight_layout()
plt.show()
````

````python
import numpy as np
import matplotlib.pyplot as plt

def six_point_averaging(x):
    len_x = len(x)
    y = np.zeros_like(x, dtype=float)
    # y = []
    for n in range (len_x):
        sum = 0
        for i in range(6):
            if(n-i>=0):
                sum+=x[n-i]
        y[n]=sum/6
        # y.append(sum/6)
    return y
def six_point_differencing(x):
    len_x = len(x)
    y = np.zeros_like(x,dtype=float)
    for n in range(5,len_x):
        y[n] = (x[n]-x[n-1]+x[n-2]-x[n-3]+x[n-4]-x[n-5])/6
    return y


x = [1,3,-2,4,5,-1,6,8,2,-3]
avg1 = six_point_averaging(x)
diff = six_point_differencing(x)
print("six point averaging",[round(float(val),2) for val in avg1])
print("six point differencing",[round(float(val),2) for val in diff])
````
